##**filling cast column missing values**
**by random technique**

```python
missing = x["cast"].isnull().sum()
```

```python
random_cast_list = x['cast'].dropna().sample(missing, replace=True).values
```

```python
# here we are just finding the info at where is my missing
# or at which index  my missing values are actually  placed

missing_indexes = x[x['cast'].isnull()].index
 # in these indexes our values are missing
 # now we will fill
```

```python
x.loc[missing_indexes,'cast'] = random_cast_list
```

```python
p = pd.read_csv("/content/titanic_toy.csv")
```

```python
p.head(1)
```

```python
p.isnull().sum()
```

```python
p['Age'].value_counts()
```

```python
m = p
```

```python
m
```

```python
missing = m["Age"].isnull().sum()
```

```python
random_Age_list = m['Age'].dropna().sample(missing, replace=True).values
```

```python
missing_indexes = m[m['Age'].isnull()].index
```

```python
 m.loc[missing_indexes,'Age'] = random_Age_list
```

```python
p.isnull().sum()
```

```python
m['Age'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

# Original Age
sns.kdeplot(p['Age'], label='Original', color='blue', fill=True, alpha=0.3)

# Imputed Age
sns.kdeplot(m['Age'], label='Imputed', color='orange', fill=True, alpha=0.3)

plt.title('Comparison of Original and Imputed Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

# Original Age - blue line
sns.kdeplot(p['Age'], label='Original', color='blue', fill=False, linewidth=2)

# Imputed Age - orange line
sns.kdeplot(m['Age'], label='Imputed', color='orange', fill=False, linewidth=2)

plt.title('Comparison of Original and Imputed Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(12,6))
sns.set_style("darkgrid")  # thoda vibrant background

# Original Age distribution with fancy color map
sns.histplot(p['Age'],
             kde=True,
             bins=30,
             color='deepskyblue',
             alpha=0.7,
             edgecolor='navy',
             linewidth=1.5,
             label='Original')

# Imputed Age distribution with contrasting vibrant color
sns.histplot(m['Age'],
             kde=True,
             bins=30,
             color='gold',
             alpha=0.5,
             edgecolor='orange',
             linewidth=1.5,
             label='Imputed')

# Sparkle effect on KDE lines
for line in plt.gca().lines:
    line.set_linewidth(3)      # thodi moti line
    line.set_alpha(0.9)        # line thodi zyada visible
    line.set_linestyle('--')   # dashed lines for style

plt.title('ðŸŒŸ Age Distribution: Original vs Imputed ðŸŒŸ', fontsize=18, fontweight='bold')
plt.xlabel('Age', fontsize=14, fontweight='bold')
plt.ylabel('Frequency', fontsize=14, fontweight='bold')
plt.legend(fontsize=13, frameon=True, shadow=True)
plt.show()
```
